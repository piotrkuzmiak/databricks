```sql
WITH RankedNames AS (
    SELECT 
        `Year`,
        `First_Name`,
        `Sex`,
        `Count`,
        ROW_NUMBER() OVER (PARTITION BY `Year`, `Sex` ORDER BY `Count` DESC) AS rank
    FROM 
        `workspace`.`tutorials`.`baby_names`
)
SELECT 
    `Year`,
    `First_Name`,
    `Sex`,
    `Count`
FROM 
    RankedNames
WHERE 
    rank = 1
ORDER BY 
    `Year`, 
    `Sex`;
```
